# Datos espaciales
Es aquel que puede ser asociado a una posicion en el espacio, ya sea en plano 2D, 3D o n-dimensional.

```python
import matplotlib.pyplot as plt 

#Simular un conjunto de datos que representen la elevacion puntual en un terrero

#crear un conjunto de datos de prueba
elevacion= np.random.randint(0, 3000, size=(100,100))
```

```python
#muestra los datos de un arreglo y cmap= muestra un mapa de color
plt.imshow(elevacion, cmap="terrain") 
plt.colorbar(label= "Altura (m)")
plt.show()
```

## Taller
Crear un ndarray de 2 dimensiones con la estatura de cada estudiante

calcular la media, mediana y desviacion estandar y graficar la data de estatura del grupo

```python
miLista= ([1.88, 1.60, 1.60, 1.60, 1.60, 1.60, 1.65, 1.59, 1.73],[1.54, 1.72, 1.74, 1.80, 1.75, 1.75, 1.69, 1.75, 1.60])
estaturas = np.array(miLista)
print(estaturas)
media= np.mean(estaturas)
print(media)
mediana= np.median(estaturas)
print(mediana)
desviacion= np.std(estaturas)
print(desviacion)
plt.imshow(estaturas, cmap="BuPu")
plt.colorbar(label="Estatura (m)")
plt.show()
```
